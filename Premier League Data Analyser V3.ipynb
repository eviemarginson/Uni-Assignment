```python
# This script is a user interface for viewing football club stats, league standings and graphs.

import pandas as pd
import matplotlib.pyplot as plt
from colorama import Fore, Style
from tabulate import tabulate
import sys
import os

# List of valid Premier League team names
VALID_TEAMS = [
    "Arsenal", "Aston Villa", "Bournemouth", "Brentford", "Brighton", "Burnley", "Chelsea", "Crystal Palace",
    "Everton", "Fulham", "Ipswich", "Leicester", "Liverpool", "Luton", "Manchester City", "Manchester United",
    "Newcastle", "Nottingham", "Sheffield United", "Southampton", "Tottenham", "West Ham", "Wolverhampton"
]

# List of available seasons
SEASONS = ['23/24', '24/25']

# Path to the Excel file containing football stats
EXCEL_FILE_PATH = "C:\\Users\\eviem\\OneDrive\\Documents\\Uni\\Module 2- Digital Maths and Programming\\Uni-Assignment\\Uni-Assignment\\Combined Football Club Stats.xlsx"

# Function to print a visual divider
def divider():
    print("-" * 30)

# Function to exit the program
def exit_program():
    print(Fore.RED + "Exiting the Premier League Data Analyser. Goodbye!" + Fore.RESET)
    sys.exit()

# Function to load the Excel file containing football club stats
def load_data():
    try:
        return pd.read_excel(EXCEL_FILE_PATH)
    except Exception as e:
        # If there is an error loading the Excel file
        print(Fore.RED + f"Error loading Excel file: {e}" + Fore.RESET)
        return None

# Function to display team stats for a selected season or seasons
def display_season_stats(team_name, season_label, filter_seasons, data):
    print(f"\nLoading stats for {Fore.RED}{team_name}{Fore.RESET} - {Fore.RED}{season_label}{Fore.RESET}")
    
    # Filter the data for the selected team and season(s)
    filtered_data = data[(data['Team'] == team_name) & (data['Season'].isin(filter_seasons))]

    # If there is no data for the season(s)
    if filtered_data.empty:
        print(Fore.RED + f"No data found for {team_name} in selected season(s)." + Fore.RESET)
    # If data is found, display it in a formatted table
    else:
        print(tabulate(filtered_data.reset_index(drop=True), headers='keys', tablefmt='grid'))

# Team stats season viewer function
def team_stats_season_viewer(team_name):
    # Load the Excel file
    data = load_data()
    if data is None:
        return

    while True:
        divider()
        # Season selection menu
        print(f"Which season do you want to view stats for {Fore.RED}{team_name}{Fore.RESET}?")
        print("1. All Seasons")
        print("2. 24/25")
        print("3. 23/24")
        print("4. Return to Team Stats Viewer")
        print("5. Return to Main Menu")
        choice = input("Enter your choice: ").strip()

        # If the user selects 1 - All Seasons
        if choice == "1":
            display_season_stats(team_name, "All Seasons", SEASONS, data)
        # If the user selects 2 - 24/25 season
        elif choice == "2":
            display_season_stats(team_name, "24/25", ['24/25'], data)
        # If the user selects 3 - 23/24 season
        elif choice == "3":
            display_season_stats(team_name, "23/24", ['23/24'], data)
        # If the user selects 4 - Return to team selection
        elif choice == "4":
            return
        # If the user selects 5 - Return to main menu
        elif choice == "5":
            break
        # If the user enters an invalid choice
        else:
            print(Fore.RED + "Invalid choice. Please try again." + Fore.RESET)

        # Prompt for next navigation step
        if not handle_navigation():
            exit_program()

# Navigation options function
def handle_navigation():
    divider()
    print("Navigation Options:")
    print("1. Return to Team Stats Viewer")
    print("2. Return to Main Menu")
    print("3. Exit")
    nav_choice = input("Enter your choice: ").strip()

    # If the user selects 1 - Return to Team Stats Viewer
    if nav_choice == "1":
        divider()
        print(Fore.RED + "Returning to Team Stats Viewer..." + Fore.RESET)
        return True
    # If the user selects 2 - Return to main menu
    elif nav_choice == "2":
        divider()
        print(Fore.RED + "Returning to Main Menu..." + Fore.RESET)
        return False
    # If the user selects 3 - Exit the program
    elif nav_choice == "3":
        exit_program()
    # If the user enters an invalid choice
    else:
        divider()
        print(Fore.RED + "Invalid choice. Returning to Team Stats Viewer." + Fore.RESET)
        return True

# Team stats viewer function
def team_stats_viewer():
    while True:
        divider()
        print(Fore.RED + "Welcome to the Team Stats Viewer!" + Fore.RESET)
        print("Enter the name of the Premier League team:")
        team_name = input("Team Name: ").strip()

        # If the entered team name is valid
        if team_name in VALID_TEAMS:
            print(f"You selected: {Fore.RED}{team_name}{Fore.RESET}")
            # Show season selection for the team
            team_stats_season_viewer(team_name)
        else:
            # If the entered team is invalid
            print(Fore.RED + f"No data found for '{team_name}'." + Fore.RESET)
            see_list = input("Would you like to see a list of valid teams? (y/n): ").strip().lower()
            if see_list == "y":
                print("\n".join(VALID_TEAMS))

# Placeholder for league standings (not yet implemented)
def display_league_standings():
    print(Fore.RED + "League Standings Viewer is not implemented yet." + Fore.RESET)

# Placeholder for graph viewing (not yet implemented)
def display_graphs():
    print(Fore.RED + "Graph Viewer is not yet implemented." + Fore.RESET)

# Main function to display main menu and handle user navigation
def main():
    while True:
        divider()
        print("Welcome to the Premier League Data Analyser!")
        print("Main Menu")
        print("1. View Team Stats")
        print("2. View League Standings")
        print("3. View Graphs")
        print("4. Exit")
        choice = input("Enter your choice: ").strip()

        # If the user selects 1 - View Team Stats
        if choice == "1":
            team_stats_viewer()
        # If the user selects 2 - View League Standings
        elif choice == "2":
            display_league_standings()
        # If the user selects 3 - View Graphs
        elif choice == "3":
            display_graphs()
        # If the user selects 4 - Exit
        elif choice == "4":
            exit_program()
        # If the user enters an invalid choice
        else:
            print(Fore.RED + "Invalid input. Please try again." + Fore.RESET)

# Entry point for the program
if __name__ == "__main__":
    main()
```
